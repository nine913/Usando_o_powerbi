```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando dados
base = {
    'Produto': ['Produto A', 'Produto B', 'Produto C', 'Produto D'],
    'Vendas': [100, 200, 300, 400]
}
base = pd.DataFrame(base)

# Criando o gráfico
sns.barplot(x='Produto', y='Vendas', data=base)

# Exibe o gráfico
plt.show()
```
